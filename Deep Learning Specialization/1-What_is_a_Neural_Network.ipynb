## Example: Housing Price Prediction
- Input: size
- Output: price

![Size-Price Graph](images/size_price_graph.png)

As shown in the graph, when the size falls below a certain point, the price stays at 0. Since price cannot be negative. The graph looks like it is <b>ReLU</b>, standing for Rectified Linear Unit.

![Single Neuron](images/single_neuron.png)

This is a single neuron. And a larger neural network is then formed by many of single neurons. It generate the function shown in the blue graph.

Suppose size, the number of bedrooms, zip code, and wealth are used to predict family size, walkability, and school quality. And these three predict the most important value, price.

![Price Prediction Network](images/price_predict_network.png)

- Input: size, #bedrooms, zip code, wealth
- Output: price

This network is an assembly of individual neurons.

Hidden layers are fully connected to all inputs, and you don't need to choose which nodes feed into the hidden layer. Inputs with less relevance are automatically adjusted to have small 'weight'.
